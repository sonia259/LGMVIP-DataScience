### Converting original image into pencil sketch

### IMPORT STATEMENTS

```python
import cv2
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display
```

### Read Image and Display

```python
img = cv2.imread('pic.jpg',1)
img.shape
```

```python
cv2.imshow('Image',img)  # opens a pop up 
cv2.waitKey(0)  # This will close one particular image opened
cv2.destroyAllWindows()  # Closes all the image pop ups opened all at time
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img1))
```

### Converting the image into Grayscale image

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale Image", gray_image)
cv2.waitKey(0)
```

```python
display(Image.fromarray(gray_image))
```

```python
# invert the grayscale image
inverted_image = 255 - gray_image
cv2.imshow("Inverted Image", inverted_image)
cv2.waitKey()
```

```python
# display(Image.fromarray(inverted_image))
```

```python
# blur the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
cv2.imshow("original image", img)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```
